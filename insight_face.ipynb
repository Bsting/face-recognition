```python
import mtcnn
import numpy as np
import math
import time

from faceEngine import FaceEngine
from mtcnn import MTCNN
from PIL import Image, ImageDraw

mtcnn = MTCNN()
faceEngine = FaceEngine()
faceEngine.load_state('pretrained_model/model.pth')
```

```python
def test_face_engine(subject_file, target_file):    
    subject_image = Image.open(subject_file)
    target_image = Image.open(target_file)
    
    start_time = time.time()
    bbox, face, _ = mtcnn.align_largest_bbox(subject_image)
    print(f'Subject face detect time: {int((time.time() - start_time) * 1000)} ms')
    
    start_time = time.time()
    subject_emb = faceEngine.extract_single(face)
    print(f'Subject face extract time: {int((time.time() - start_time) * 1000)} ms')
    
    start_time = time.time()
    box, face, _ = mtcnn.align_largest_bbox(target_image)
    print(f'Target face detect time: {int((time.time() - start_time) * 1000)} ms')
    
    start_time = time.time()
    target_emb = faceEngine.extract_single(face)
    print(f'Target face extract time: {int((time.time() - start_time) * 1000)} ms')
    
    diff = target_emb - subject_emb
    dist = np.sum(np.power(diff, 2))
    score = (1.0 - (dist/(2*math.pi))) * 100

    print(f'Dist: {dist}, score: {score}')
```

```python
file_1 = './data/test/jay_1.jpg'
file_2 = './data/test/jay_2.jpg'
test_face_engine(file_1, file_2)
```
